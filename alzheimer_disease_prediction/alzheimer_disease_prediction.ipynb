```python
data.drop(['DoctorInCharge', 'PatientID'], axis=1, inplace=True)
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 20))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
columns = ['Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 'SleepQuality', 'SystolicBP',
           'DiastolicBP', 'CholesterolTotal', 'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides', 'MMSE',
           'FunctionalAssessment', 'ADL']

figs, axes = plt.subplots(3, 5, figsize=(20, 15))
axes = axes.flatten()

for i, col in enumerate(columns):
    axes[i] = sns.histplot(data=data, x=col, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
columns = data.drop(
    ['Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 'SleepQuality', 'SystolicBP', 'DiastolicBP',
     'CholesterolTotal', 'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides', 'MMSE', 'FunctionalAssessment',
     'ADL'], axis=1).columns

figs, axes = plt.subplots(3, 6, figsize=(25, 20))
axes = axes.flatten()

for i, col in enumerate(columns):
    sns.countplot(data=data, x=col, hue=col, ax=axes[i], palette="Set2", legend=False)
    axes[i].set_title(f'Count Plot of {col}')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder

x_data = data.drop('Diagnosis', axis=1)
y_data = data['Diagnosis']

# One-hot encoding
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
categorical_features = data.drop(
    ['Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 'SleepQuality', 'SystolicBP', 'DiastolicBP',
     'CholesterolTotal', 'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides', 'MMSE', 'FunctionalAssessment',
     'ADL', 'Diagnosis'], axis=1).columns
original = data.drop('Diagnosis', axis=1)
encoded = one_hot_encoder.fit_transform(original[categorical_features])
train_df = pd.DataFrame(encoded, columns=one_hot_encoder.get_feature_names_out(categorical_features))
numerical_features = original.drop(categorical_features, axis=1)
x_data = pd.concat([numerical_features, train_df], axis=1)

# Standardize the data after one-hot encoding
x_data = StandardScaler().fit_transform(x_data)

# Split train set and evaluation set
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Giả sử y_data là mảng numpy với các nhãn (số nguyên) của từng class
unique_classes = np.unique(y_data)
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_classes)))

# PCA
pca = PCA(n_components=3, random_state=42)
x_data_pca = pca.fit_transform(x_data)

# Visualization
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')

for i, class_label in enumerate(unique_classes):
    indices = np.where(y_data == class_label)
    ax.scatter(
        x_data_pca[indices, 0],
        x_data_pca[indices, 1],
        x_data_pca[indices, 2],
        color=colors[i],
        label=f'Class {class_label}',
    )

ax.set_title('Data Visualization (PCA 3D)')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.legend()  # Thêm chú thích
plt.show()
```

```python
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Naive Bayes Classifier
naive_bayes = GaussianNB()

start_time = time.perf_counter()
naive_bayes.fit(x_train, y_train)
end_time = time.perf_counter()

# Evaluate model
y_pred_naive = naive_bayes.predict(x_test)
print(f"Training time: {end_time - start_time} seconds")
print(f"Accuracy: {accuracy_score(y_test, y_pred_naive)}")
print(classification_report(y_test, y_pred_naive))
```

```python
from sklearn.linear_model import LogisticRegression

# Logistic Regression
logistic = LogisticRegression(solver='saga')

start_time = time.perf_counter()
logistic.fit(x_train, y_train)
end_time = time.perf_counter()

# Evaluate model
y_pred_log = logistic.predict(x_test)
print(f"Training time: {end_time - start_time} seconds")
print(f"Accuracy: {accuracy_score(y_test, y_pred_log)}")
print(classification_report(y_test, y_pred_log))
```

```python
from sklearn.svm import SVC

# Support Vetor Machine
svm = SVC(kernel='linear', C=1)  # Soft Margins

start_time = time.perf_counter()
svm.fit(x_train, y_train)
end_time = time.perf_counter()

# Evaluate model
y_pred_svm = svm.predict(x_test)
print(f"Training time: {end_time - start_time} seconds")
print(f"Accuracy: {accuracy_score(y_test, y_pred_svm)}")
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)

start_time = time.perf_counter()
random_forest.fit(x_train, y_train)
end_time = time.perf_counter()

# Evaluate model
y_pred = random_forest.predict(x_test)
print(f"Training time: {end_time - start_time} seconds")
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(classification_report(y_test, y_pred))
```
